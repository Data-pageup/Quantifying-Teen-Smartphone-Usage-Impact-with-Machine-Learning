# importing the library 

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

```

# importing the dataset 

```python
df = pd.read_csv(r"E:\Machine learning project\Teen Smartphone Usage and Addiction Impact Dataset\teen_phone_addiction_dataset.csv")
```

```python
df.shape
```

```python
df.head(10)
```

# data- preprocessing

### dealing with categorical values

```python
df.select_dtypes(include =[ 'object', 'category'])
```

```python
# names we can drop 
df.drop(columns = ['Name'], inplace = True )
```

```python
df.drop(columns = ['ID'], inplace = True )
```

```python
df.Gender.unique()  # 0 1 2
```

```python
df.Location.nunique()  # we are droping 
```

```python
df.drop(columns = ['Location'], inplace = True )
```

```python
df.School_Grade.unique()   # 7 8 9 10 11 12
```

```python
df.Phone_Usage_Purpose.unique()  #one -hot encoding (creating multiple columns )
```

## Now mapping gender, school grade, phone usage purpose

```python
gender_map = { 0:'Male',1:'Female',2:'other' }
df['Gender'] = df['Gender'].map(gender_map)
```

```python
School_Grade_map = {'7th':7 , '8th':8,'9th':9,'10th':10,'11th':11,'12th':12}
df['School_Grade'] = df['School_Grade'].map(School_Grade_map)
```

```python
df = pd.get_dummies(df, columns = ['Phone_Usage_Purpose'], prefix = "Purpose")
```

```python
df.columns 
```

```python
df.select_dtypes(include = ["int",'float'])
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.Gender
```

```python
# map text labels to integers
str_to_code = {"Female": 0, "Male": 1, "Other": 2}
df["Gender_Code"] = df["Gender"].map(str_to_code)
```

```python
df.Gender_Code
```

```python
df.drop(columns = ['Gender'], inplace = True )
```

```python
x = df.drop(columns = ['Addiction_Level'])
```

```python
y= df.Addiction_Level
```

# train test split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, random_state = 42, test_size = 0.3)
```

```python
model1 = LinearRegression()
```

```python
model1.fit(x_train,y_train)
```

```python
y_pred = model1.predict(x_test)
```

## evaluating 

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
print(mean_absolute_error(y_test,y_pred))
```

```python
print(mean_squared_error(y_test,y_pred))
```

```python
print(r2_score(y_test,y_pred))
```

```python
from sklearn.linear_model import Ridge,Lasso
from sklearn.tree import DecisionTreeRegressor
```

# Ridge

```python
model2 = Ridge(alpha = 0.5)
```

```python
model2.fit(x_train,y_train)
```

```python
y_pred1 = model2.predict(x_test)
print(mean_absolute_error(y_test,y_pred1))
print(mean_squared_error(y_test,y_pred1))
print(r2_score(y_test,y_pred1))
```

## Lasso

```python
model3 = Lasso(alpha = 0.05)
```

```python
model3.fit(x_train,y_train)
```

```python
y_pred2 = model3.predict(x_test)
print(mean_absolute_error(y_test,y_pred2))
print(mean_squared_error(y_test,y_pred2))
print(r2_score(y_test,y_pred2))
```

# Decision Tree 

```python
model4 = DecisionTreeRegressor(max_depth = 100)
```

```python
model4.fit(x_train,y_train)
```

```python
y_pred3 = model4.predict(x_test)
print(mean_absolute_error(y_test,y_pred3))
print(mean_squared_error(y_test,y_pred3))
print(r2_score(y_test,y_pred3))
```
